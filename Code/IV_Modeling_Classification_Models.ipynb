```python
df =  pd.read_csv('../Data/df_clean_final.csv')
```

```python
df.Rating.value_counts()
```

```python
#Binarize the Predictor Variable: Satisfaction Rating 
# 1: Not Satisfied with Rental (Satisfaction Rating of 1,2,3) 
# 0: Satisfied with Rental (Satisfaction Rating 4,5)
df['Rating'] = np.where(df['Rating'] <= 3, 1, 0)
```

```python
#Drop Unnecessary Columns 
df.dropna(inplace=True)
df = df.reset_index()
df.drop(columns="index",inplace=True)
```

```python
X = df.drop(columns=['Rating'])
y = df['Rating']
```

```python
def get_categorical_indicies(X):
    cats = []
    for col in X.columns:
        if pd.api.types.is_numeric_dtype(X[col]):
            pass
        else:
            cats.append(col)
    cat_indicies = []
    for col in cats:
        cat_indicies.append(X.columns.get_loc(col))
    return cat_indicies
```

```python
def convert_cats(X):
    cats = []
    for col in X.columns:
        if pd.api.types.is_numeric_dtype(X[col]):
            pass
    else:
        cats.append(col)
        cat_indicies = []
    for col in cats:
        X[col] = X[col].astype('category')
```

```python
categorical_indicies = get_categorical_indicies(X)
```

```python
convert_cats(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101, stratify=y)
```

```python
y.value_counts(normalize=True)
```

```python
train_df = pd.concat([X,y],axis=1)
nothappy= train_df[train_df['Rating']==1]
happy = train_df[train_df['Rating']==0]
happy = happy.sample(n=len(nothappy), random_state=101)
train_df = pd.concat([nothappy,happy],axis=0)
X_train = train_df.drop('Rating',axis=1)
y_train = train_df['Rating']
```

```python
train_dataset = cb.Pool(X_train,y_train, cat_features=categorical_indicies)
test_dataset = cb.Pool(X_test,y_test, cat_features=categorical_indicies)
```

```python
model_AUC = cb.CatBoostClassifier(loss_function='Logloss', eval_metric='AUC')
```

```python
grid_AUC = {'learning_rate': [0.03, 0.1],
'depth': [4, 6, 10],
'l2_leaf_reg': [1, 3, 5,],
'iterations': [50, 100, 150]}
```

```python
model_AUC.grid_search(grid_AUC,train_dataset)
```

```python
pred = model_AUC.predict(X_test)
print(classification_report(y_test, pred))
```
